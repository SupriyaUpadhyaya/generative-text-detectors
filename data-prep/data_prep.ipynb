## Remove the unicodes and escape characters from wiki data

```python
import os
import json

def process_text(text):
    """
    This function takes a text string and processes it by decoding unicode
    characters and replacing escape characters with their actual representations.
    """
    return text.encode('utf-8').decode('unicode_escape').encode('latin1').decode('utf-8')

def process_jsonl_files(input_dir, output_dir):
    """
    This function reads all the JSONL files from the input directory,
    processes each line by replacing the values of 'text' and 'machine_abstract'
    with their unicode-decoded and escape-character-replaced versions,
    and saves the processed content to new JSONL files in the output directory.
    """
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for filename in os.listdir(input_dir):
        if filename.endswith(".jsonl"):
            print("filename : ", filename)
            input_path = os.path.join(input_dir, filename)
            output_path = os.path.join(output_dir, filename)

            with open(input_path, 'r', encoding='utf-8') as infile, open(output_path, 'w', encoding='utf-8') as outfile:
                for line in infile:
                    data = json.loads(line)
                    if 'text' in data:
                        data['text'] = process_text(data['text'])
                    if 'machine_abstract' in data:
                        data['machine_abstract'] = process_text(data['machine_abstract'])
                    if 'human_text' in data:
                        data['human_text'] = process_text(data['human_text'])
                    outfile.write(json.dumps(data, ensure_ascii=False) + '\n')

# Example usage
input_directory = '/Users/supriyaupadhyaya/ovgu/generative-text-detectors/m4/wiki/'
output_directory = '/Users/supriyaupadhyaya/ovgu/generative-text-detectors/m4/wiki_processed/'
process_jsonl_files(input_directory, output_directory)
```

## split wiki data into training, validation and test

```python
import os
import json
import random
from math import ceil

def split_data(data, train_ratio=0.64, test_ratio=0.20, val_ratio=0.16, seed=3):
    """
    Split data into train, test, and validation sets.
    """
    total_len = len(data)
    train_len = ceil(total_len * train_ratio)
    test_len = ceil(total_len * test_ratio)
    val_len = total_len - train_len - test_len

    random.seed(seed)
    random.shuffle(data)
    train_data = data[:train_len]
    test_data = data[train_len:train_len + test_len]
    val_data = data[train_len + test_len:]

    return train_data, test_data, val_data

def process_jsonl_files(input_dir, output_dir, seed=3):
    """
    Read all the JSONL files from the input directory, shuffle and split each into
    train, test, and validation sets, and save them in the output directory.
    """
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    for filename in os.listdir(input_dir):
        if filename.endswith(".jsonl"):
            input_path = os.path.join(input_dir, filename)

            with open(input_path, 'r', encoding='utf-8') as infile:
                data = [json.loads(line) for line in infile]

            train_data, test_data, val_data = split_data(data, seed=seed)

            base_filename = os.path.splitext(filename)[0]
            train_output_path = os.path.join(output_dir, f"{base_filename}_train.jsonl")
            test_output_path = os.path.join(output_dir, f"{base_filename}_test.jsonl")
            val_output_path = os.path.join(output_dir, f"{base_filename}_val.jsonl")

            with open(train_output_path, 'w', encoding='utf-8') as train_file:
                for item in train_data:
                    train_file.write(json.dumps(item, ensure_ascii=False) + '\n')

            with open(test_output_path, 'w', encoding='utf-8') as test_file:
                for item in test_data:
                    test_file.write(json.dumps(item, ensure_ascii=False) + '\n')

            with open(val_output_path, 'w', encoding='utf-8') as val_file:
                for item in val_data:
                    val_file.write(json.dumps(item, ensure_ascii=False) + '\n')

# Example usage
input_directory = '/Users/supriyaupadhyaya/ovgu/generative-text-detectors/m4/wiki_processed/'
output_directory = '/Users/supriyaupadhyaya/ovgu/generative-text-detectors/m4/wiki_processed/'
process_jsonl_files(input_directory, output_directory)
```
