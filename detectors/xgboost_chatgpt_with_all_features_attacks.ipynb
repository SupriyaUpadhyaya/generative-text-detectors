# XGBoost classifier trained and tested on M4 chatGPT

```python
def xgboost_train(X_train, y_train, X_val, y_val, X_test, y_test):
    plt.figure(figsize=(10, 6))
    lr = 0.01
    weight = 4.5
    evalset = [(X_train, y_train), (X_val,y_val)]

    importances_gain = pd.DataFrame()
    importances_weight = pd.DataFrame()
    importances_cover = pd.DataFrame()
    importances_total_gain = pd.DataFrame()

    xgb_classifier = XGBClassifier(n_estimators=500,
                                use_label_encoder=False,
                                eval_metric="logloss",
                                early_stopping_rounds=5,
                                n_jobs=-1,
                                eta=lr,
                                reg_lambda=1,
                                min_child_weight=weight)
    xgb_classifier.fit(X_train, y_train,
                    eval_set=evalset,
                    verbose=False)
    results = xgb_classifier.evals_result()
    yhat = xgb_classifier.predict(X_test)
    score = accuracy_score(y_test, yhat)
    print('Accuracy: %.3f' % score)

    plt.plot(results['validation_0']['logloss'], label='train')
    plt.plot(results['validation_1']['logloss'], label='validation')
    values = list(X_train.columns.values)
    title = ' '.join(values)
    plt.title('Loss vs. Epoch for ')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    # show the legend
    plt.legend()
    # show the plot
    plt.show()

    booster = xgb_classifier.get_booster()
    importance_score = booster.get_score(importance_type='gain')
    importance_frame = pd.DataFrame({'Importance': list(importance_score.values()), 
                                    'Feature': list(importance_score.keys())})
    importances_gain = pd.concat([importances_gain, 
                                 importance_frame],
                                 axis=0,
                                 sort=False)
    
    importance_score = booster.get_score(importance_type='weight')
    importance_frame = pd.DataFrame({'Importance': list(importance_score.values()), 'Feature': list(importance_score.keys())})
    importances_weight = pd.concat([importances_weight, importance_frame], axis=0, sort=False)
    
    importance_score = booster.get_score(importance_type='cover')
    importance_frame = pd.DataFrame({'Importance': list(importance_score.values()), 'Feature': list(importance_score.keys())})
    importances_cover = pd.concat([importances_cover, importance_frame], axis=0, sort=False)
    
    importance_score = booster.get_score(importance_type='total_gain')
    importance_frame = pd.DataFrame({'Importance': list(importance_score.values()), 'Feature': list(importance_score.keys())})
    importances_total_gain = pd.concat([importances_total_gain, importance_frame], axis=0, sort=False)

    mean_total_gain = importances_total_gain[['Importance', 'Feature']].groupby('Feature').mean()
    mean_total_gain = mean_total_gain.reset_index()
    plt.figure(figsize=(17, 8))
    sns.barplot(x='Importance', y='Feature', 
                                data=mean_total_gain.sort_values('Importance',
                                ascending=False), 
                                palette='gray')
    
    return xgb_classifier, score
```

```python
def performance_test(xgb_classifier, path):
    X_test_list, y_test_list, test = get_dataset(path)
    y_pred = xgb_classifier.predict(X_test_list[0])
    

    # Compute confusion matrix
    cm = confusion_matrix(y_test_list[0], y_pred)

    # Plot confusion matrix using seaborn heatmap
    plt.figure(figsize=(8, 6))
    cmn = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
    sns.heatmap(cmn, annot=True, fmt='.2f', xticklabels=['Human', 'Machine'], yticklabels=['Human', 'Machine'])
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.show(block=False)
    explainer = shap.Explainer(xgb_classifier)
    shap_values = explainer(X_test_list[0])
    shap.summary_plot(shap_values, X_test_list[0])
    shap.plots.heatmap(shap_values)

    f1score = f1_score(y_test_list[0], y_pred, zero_division=1.0)
    print("F1 score : ", f1score)
```

```python
import pandas as pd
import numpy as np

def get_text(filepath):
    df = pd.read_csv(filepath)
    #print(df.head())
    X_list = []
    y_list = []
    #human = ['human_text']
    machine = ['machine_text']

    #df_human = df.loc[:, human]
    #df_human['class'] = 0
    df_machine = df.loc[:, machine]
    df_machine['class'] = 1

    #human = {'human_text': 'text'}
    machine = {'machine_text': 'text'}

    #df_human = df_human.rename(columns=human)
    df_machine = df_machine.rename(columns=machine)

    full_df = pd.concat([df_machine], ignore_index=True)
    
    X = full_df.iloc[:,:-1]
    #print(X.head())
    y = full_df.iloc[:,-1]

    return X, y
```

```python
test_path = '/Users/supriyaupadhyaya/ovgu/generative-text-detectors/feature-extractor/features/m4_withoutN/arxiv_chatGPT_test_withoutN_feature.csv'
X_test, y_test= get_text(test_path)
X_test = X_test['text'].tolist()
y_test = y_test.tolist()
```

```python
import numpy as np
from textattack.models.wrappers import ModelWrapper
from textattack.datasets import Dataset

dataset = Dataset(X_test, y_test)
```

```python
import numpy as np
from textattack.models.wrappers import ModelWrapper
from FeatureExtractor import FeatureExtractor

class XGBoostWrapper(ModelWrapper):
    def __init__(self, model):
        self.model = model
        self.featureExtractor = FeatureExtractor()
        print("Hello model!")
        
    
    def __call__(self, text_input_list):
        print(text_input_list)
        input_array = np.array([self.featureExtractor.getFeatures(text) for text in text_input_list])
        print("Input features : ", input_array)
        probs = self.model.predict_proba(input_array)
        return probs
```

```python
import textattack
from textattack.attack_recipes import PWWSRen2019, DeepWordBugGao2018, TextFoolerJin2019
from textattack import Attacker, AttackArgs

# Wrap the XGBoost model
wrapped_model = XGBoostWrapper(xgb_pairwise_classifiers[0])

# Set up the attack arguments
attack_args = AttackArgs(num_examples=5, log_to_csv="/Users/supriyaupadhyaya/ovgu/generative-text-detectors/logs/attack_log.csv", disable_stdout=False)

# Create and run attacks
attack_methods = [PWWSRen2019, DeepWordBugGao2018, TextFoolerJin2019]
attack_names = ["PWWS", "DeepWordBug", "TextFooler"]
results = {}

#for attack_method, attack_name in zip(attack_methods, attack_names):

print("step 0")
attack = TextFoolerJin2019.build(wrapped_model)
print("step 1")
attacker = Attacker(attack, dataset, attack_args)
print("step 2")
results['TextFoolerJin2019'] = attacker.attack_dataset()
print("step 3")

# Print results
for attack_name in results:
    print(f"Results for {attack_name}:")
    for result in results[attack_name]:
        print(result)
```

```python
import matplotlib.pyplot as plt

# Count successful attacks
success_counts = {attack_name: sum(1 for result in attack_results if result.goal_status == textattack.shared.AttackGoalStatus.SUCCEEDED)
                  for attack_name, attack_results in results.items()}

# Visualize the results
attack_names = list(success_counts.keys())
successes = list(success_counts.values())

plt.figure(figsize=(10, 6))
plt.bar(attack_names, successes, color=['blue', 'green', 'red'])
plt.xlabel('Attack Methods')
plt.ylabel('Number of Successful Attacks')
plt.title('Comparison of Attack Methods')
plt.show()
```
